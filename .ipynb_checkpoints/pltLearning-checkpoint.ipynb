```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,50)
y1 = x**2 +1
y2 = 2*x + 1
```

```python
plt.figure()
plt.plot(x,y1)
plt.show()
```

```python
plt.figure(num=3,figsize=(8,5))
plt.plot(x,y2)
plt.plot(x,y1,color = 'red',linewidth = 1.0,linestyle = '--')
plt.show()
```
